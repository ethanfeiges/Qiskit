```python
# Ensure that plots are displayed inline in Jupyter notebooks
%matplotlib inline

# Import necessary libraries
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_histogram, plot_bloch_multivector
import matplotlib.pyplot as plt




```

```python
qc = QuantumCircuit(1)
# Apply Hadamard gate
qc.h(0)
qc.h(0)
```

```python

# Draw the circuit
qc.draw(output='mpl')
plt.show()  # Ensure the circuit diagram is displayed within the notebook
```

```python
# Get the statevector of the circuit
statevector = Statevector.from_instruction(qc)

# Plot the Bloch spheres
plot_bloch_multivector(statevector)
plt.show()  # Ensure the Bloch spheres plot is displayed within the notebook

```

```python
# Simulate measurements
# Simulate measuring the qubits 1024 times
counts = statevector.sample_counts(shots=1024)

# Plot the histogram of the measurement outcomes
plot_histogram(counts)
plt.show()  # Ensure the histogram plot is displayed within the notebook
```
